```python
import pandas as pd
import os
import numpy as np
```

```python
f1_master = pd.DataFrame()
for config in os.listdir("europe"):
    path = "europe/"+config+"/results.csv"
    if os.path.isfile(path):
        temp = pd.read_csv(path)
        temp.set_index("country", inplace=True)
        temp.sort_index(inplace=True)
        f1_master[config] = temp.f1

# Drop very small countries (AND, LIE)
f1_master.drop(['AND', 'LIE'], inplace=True)
```

```python
recall_master = pd.DataFrame()
config_lists = os.listdir("europe")
config_lists.sort()
for config in config_lists:
    path = "europe/"+config+"/results.csv"
    if os.path.isfile(path):
        temp = pd.read_csv(path)
        temp.set_index("country", inplace=True)
        temp.sort_index(inplace=True)
        recall_master[config] = temp.recall
        recall_master.head()
        #recall_master[config] = recall_master[config]-recall_master[baseline]

# Drop very small countries (AND, LIE)
for config in config_lists[3:]:
    recall_master[config] = recall_master[config]-recall_master["baseline"]
recall_master.drop(['AND', 'LIE'], inplace=True)
recall_master
#.mean(axis=0)
```

```python
# h0 my1 = 0.43
# h1 my2 > 0.43
# alpha = 0.05
baseline_stats = f1_master.baseline.describe()
baseline_stats
```

$T = \frac{\overline{x_{1}}-\overline{x_{2}}}{\frac{s_{diff}}{\sqrt{n}}}$

```python

base = f1_master.baseline
test = f1_master["pa1-s5"]

base_mean = base.mean()
test_mean = test.mean()
diff = base-test
s_diff = diff.std()
n = base.count()
print("Base mean:", base_mean,"\nTest mean:", test_mean, "\nStd of differences:", s_diff,"\nCount:", n)

T = (base_mean-test_mean)/(s_diff/np.sqrt(n))
print("T-value:", T.round(2))
```
