```python
import os
import pandas as pd
import operator
```

```python
results_path = "Data/results/europe/"
scores = {}
for threshold in os.listdir(results_path):
    if "base" in threshold:
        results = pd.read_csv(results_path+"/"+threshold+"/results.csv")
        scores[threshold] = results.f1.mean()
    

best = max(scores.items(), key=operator.itemgetter(1))[0]
print(best, scores[best])
#scores = pd.DataFrame.from_dict(scores, orient="index")
#scores.sort_values(by=0)
scores
```

```python
filtered_path = "Data/results/europe/lccs_filtered/results.csv"
filt = pd.read_csv(filtered_path)
filt_results = filt.f1.mean()
print(best, round(scores[best], 4), "vs. Waterfiltered", round(filt_results, 4))
```

```python
lccs_path = "Data/results/europe/lccs_filtered/results.csv"
pop1_lccs_path = "Data/results/europe/pop1_lccs/results.csv"
pop10_lccs_path = "Data/results/europe/pop10_lccs/results.csv"
slope_pop1_path = "Data/results/europe/slope_pop1_lccs/results.csv"

lccs = pd.read_csv(lccs_path)
pop1 = pd.read_csv(pop1_lccs_path)
pop10 = pd.read_csv(pop10_lccs_path)
slope_pop1 = pd.read_csv(slope_pop1_path)

lccs_results = lccs.f1.mean()
pop1_results = pop1.f1.mean()
pop10_results = pop10.f1.mean()
slope_pop1_results = slope_pop1.f1.mean()

print(best, round(scores[best], 4), "\nLCCS", round(filt_results, 4), "\nPOP1", pop1_results, "\nPOP10", pop10_results, "\nSlope", slope_pop1_results)

lccs = lccs.sort_values("country")
lccs = lccs.set_index("country")

pop1 = pop1.sort_values("country")
pop1 = pop1.set_index("country")

pop10 = pop10.sort_values("country")
pop10 = pop10.set_index("country") 

slope = slope_pop1.sort_values("country")
slope = slope_pop1.set_index("country")
```
